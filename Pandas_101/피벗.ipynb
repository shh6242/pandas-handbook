# 피벗 df.pivot()

- Reshaping 데이터 테이블 재배치
- 여러 column을 index, columns, value argument로 지정
- GroupBy 연산, 테이블 요약, 그래프 등을 위해 사용
- set_index()로 MultiIndex 생성후, unstack() 메서드로 형태를 변경하는 과정을 단축시킴

p1.pivot(index=None, columns=None, values=None) 

````python
import pandas as pd
import numpy as np
````

````python
p1 = pd.read_csv('./data/melt_ex.csv')
````

````python
p1.head()
````

````python
p1.pivot(index='year', columns='mon')
````

````python
p1.set_index(['year', 'mon']).unstack()
````

````python
p1.pivot('year', 'mon').loc[:,'latte']
````

````python
p1.pivot('year', 'mon').xs(1, level=1, axis=1) #cross-section 1월 데이터
````

````python
df1 = pd.DataFrame(np.random.randn(5,3), columns=['code', 'product', 'sales'])
df1['code'] = 'c1'
df1['product'] = ['notebook', 'notebook', 'tv', 'refrigerator', 'washer']
df1['sales'] = [250, 300, 149, 300, 120]
````

````python
df2 = pd.DataFrame(np.random.randn(4,3), columns=['code', 'product', 'sales'])
df2['code'] = 'c2'
df2['product'] = ['notebook', 'tv', 'refrigerator', 'washer']
df2['sales'] = [180, 150, 290, 220]
````

````python
df3 = pd.DataFrame(np.random.randn(4,3), columns=['code', 'product', 'sales'])
df3['code'] = 'c3'
df3['product'] = ['notebook', 'tv', 'refrigerator', 'washer']
df3['sales'] = [280, 160, 274, 140]
````

````python
data = pd.concat([df1,df2,df3])
````

````python
data.pivot_table(index='code', columns='product')
````

````python
data.pivot_table(index='code', columns='product', aggfunc=sum)
````

````python
data.pivot_table(index='code', columns='product', aggfunc=sum, margins=True)
````
