```python
import pandas as pd
import numpy as np
```

```python
mi = pd.MultiIndex.from_product([[1,2], ['s1', 's2']])
```

```python
mi_columns = pd.MultiIndex.from_product([['apple', 'banana'],
                                         ['price', 'qty']])
```

```python
df = pd.DataFrame(np.arange(len(mi)*len(mi_columns)).reshape(len(mi), len(mi_columns)), 
                  index=mi,
                  columns=mi_columns)
```

```python
df
```

# `DataFrame` Stacking

## Unstack(): 인덱스를 칼럼화 (level= -1 최하위부터, fill_value =None)

```python
df.unstack()
```

```python
df.unstack(level=0)
```

## Stack(): 컬럼을 인덱스화 (level=-1, dropna=True)

```python
df.stack()
```

```python
df.stack(level=0)
```
